### (0) Import libraries 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### (1) Definition of random number generation and limits

```python
# Parameters of the Gamma distribution

mean = 10
sigma = np.sqrt(10.0)

k = (mean / sigma) ** 2

lam = sigma**2 / mean

# Considered limits

limlow = np.array(
    [
        mean - sigma,
        mean - 2 * sigma,
        mean - 3 * sigma,
        mean - 4 * sigma,
        mean - 5 * sigma,
    ]
)
limhig = np.array(
    [
        mean + sigma,
        mean + 2 * sigma,
        mean + 3 * sigma,
        mean + 4 * sigma,
        mean + 5 * sigma,
    ]
)

# Number generation procedure


def myrnd(N=1):
    rvec = np.random.gamma(k, scale=lam, size=(N))

    return rvec
```

### (2) Generate multiple numbers

```python
Nmc = 100000000

rvec = myrnd(Nmc)
```

### (3)  Count limit probability

```python
print("\nLower limit probability:")

for ll in limlow:
    nl = np.sum(rvec < ll)
    p = nl / Nmc
    ep = np.sqrt(p * (1 - p) / Nmc)
    print(" P ( x < ", ll, ") = ", p, " +/- ", ep)

print("\nUpper limit probability:")

for ul in limhig:
    nl = np.sum(rvec > ul)
    p = nl / Nmc
    ep = np.sqrt(p * (1 - p) / Nmc)
    print(" P ( x > ", ul, ") = ", p, " +/- ", ep)
```

### (4) Plot distribution histogram

```python
# Range for result histogram

Xmax = 40.0
Nbin = 100

# Plot histogram

plt.figure(figsize=(9, 6))

plt.hist(
    rvec,
    bins=Nbin,
    range=[0.0, Xmax],
    density=True,
    log=True,
    histtype="step",
    linewidth=3,
)

plt.title("Probability distribution from MC", size=24)
plt.xlabel("variable", size=24)
plt.ylabel("Probability density", size=24)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlim([0.0, Xmax])
plt.grid(which="major", axis="both")

fname = "06_interval_3.png"
plt.savefig(fname)

plt.show()
```
